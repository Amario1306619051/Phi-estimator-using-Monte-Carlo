````python
import pygame
import random

# Initialize Pygame
pygame.init()

# Screen dimensions
screen_width = 400
screen_height = 400

# Colors
blue = (0, 0, 255)
red = (255, 0, 0)

# Create the Pygame screen
screen = pygame.display.set_mode((screen_width, screen_height))
pygame.display.set_caption("Monte Carlo Pi Estimation")

# Function to estimate pi using Monte Carlo simulation
def estimate_pi(num_points):
    points_inside_circle = 0
    
    for _ in range(num_points):
        x = random.uniform(0, 1)
        y = random.uniform(0, 1)
        
        # Calculating the distance from point (x, y) to the center of the circle (0.5, 0.5)
        distance = (x - 0.5)**2 + (y - 0.5)**2
        
        # If the distance is less than or equal to 0.25 (squared radius), then the point is inside the circle
        if distance <= 0.25:
            points_inside_circle += 1
    
    # The ratio of points inside the circle to the total points is an estimate of pi/4
    estimated_pi = 4 * points_inside_circle / num_points
    return estimated_pi

# Number of points for estimation
num_points = 100

# Estimate pi
estimated_pi = estimate_pi(num_points)
print(f"Estimated value of pi after {num_points} trials is: {estimated_pi}")

# Frame rate control
clock = pygame.time.Clock()
frame_rate = 10  # Adjust the frame rate here

# Main loop for Pygame visualization
running = True
points_plotted = 0
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
    
    # Draw the quarter circle
    pygame.draw.arc(screen, blue, (0, 0, screen_width, screen_height), 0, 90, 1)
    
    # Monte Carlo simulation
    for _ in range(num_points - points_plotted):
        x = random.uniform(0, 1) * screen_width
        y = random.uniform(0, 1) * screen_height
        
        distance = (x - screen_width / 2)**2 + (y - screen_height / 2)**2
        
        if distance <= (screen_width / 2)**2:
            pygame.draw.circle(screen, blue, (int(x), int(y)), 1)
        else:
            pygame.draw.circle(screen, red, (int(x), int(y)), 1)
    
    pygame.display.flip()  # Update the screen
    points_plotted += num_points - points_plotted
    
    clock.tick(frame_rate)  # Limit the frame rate

    # Automatically exit when all points are plotted
    if points_plotted >= num_points:
        running = False

# Quit Pygame
pygame.quit()
````
